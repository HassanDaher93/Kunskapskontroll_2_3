### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

# Svar:
* Figure - Själva rit-tavlan som skapas, innehåller en eller flera subplots.
* Axes - En plot eller graf inom rit-tavlan samt titlar, etiketter och linjer.
* Axis - Mäter data på ritningstavlan med hjälp av x-axis och y-axis.
* Artist - Allt som syns på rit-tavlan.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Data
x = np.linspace(-4, 4, 100)
y = x**2

# Figure, axes, title
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("Object-Oriented Approach")
```

```python
# Data
x = np.linspace(-4, 4, 100)
y = x**2

# Figure, axes, title
plt.plot(x, y)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Pyplot Approach")
```

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
colors = ["purple", "red", "orange", "yellow", "green"]

# Figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), layout="constrained")

# Scatter plot
ax1.scatter(random_data_x, random_data_y)
ax1.set_title("Scatter plot")

# Bar plot
ax2.bar(names, values, color=colors)
ax2.set_title("Bar plot")
```
